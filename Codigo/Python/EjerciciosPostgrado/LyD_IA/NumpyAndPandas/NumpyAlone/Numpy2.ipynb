```python

# path to the dataset
filepath = '../../datasets/MilitarySpending.csv'

# csv with numpy
headers = np.genfromtxt(filepath, delimiter=',', max_rows=1, dtype=str)
data = np.genfromtxt(filepath, delimiter=',', skip_header=1)

# Print the first 5 rows of the dataset
print(headers)
print(data[:5])
```
