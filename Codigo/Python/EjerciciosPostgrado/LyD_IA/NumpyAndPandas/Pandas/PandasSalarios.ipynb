```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('../../datasets/salary_data.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#removing the null values
df = df.dropna()
df.isnull().sum()
```

```python
df['Salary'].plot(kind='box')
```

```python
df['Years of Experience'].plot(kind='box')
```

```python
df.plot(kind='scatter', x='Years of Experience', y='Salary')
```

```python
df.plot(kind='hist', y='Salary', bins=30)
```

```python
df.groupby('Gender')['Salary'].mean().plot(kind='bar')
```

```python
df.groupby('Gender')['Years of Experience'].mean().plot(kind='bar')
```

```python
df.groupby('Gender')['Years of Experience'].agg(['mean', 'median', 'std', 'max', 'min'])
```

```python

genders = ['Female', 'Male']
for gender in genders:
    df[df['Gender'] == gender].plot(kind='scatter', x='Years of Experience', y='Salary', title=gender)
    plt.ylim(0, 250000)
    plt.xlim(0, 35)
    plt.show()
```
